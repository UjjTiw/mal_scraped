```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import timeit
```

```python
def get_webpages(n):
    webpages = []
    link = 'https://myanimelist.net/topanime.php'
    for _ in range(n//50):
        if _ == 0:
            webpages.append(link)
        else:
            num_suffix = _ * 50
            webpages.append(link + '?limit=' + str(num_suffix))
    return webpages
```

```python
def get_anime_links(webpage_link):
    anime_web = requests.get(webpage_link).text
    soup = BeautifulSoup(anime_web, 'lxml')
    anime_cards = soup.find_all('tr', class_ = 'ranking-list')
    anime_linklist = []
    for anime_card in anime_cards:
        anime_linklist.append(anime_card.find('td', class_= 'title al va-t word-break').find('a',href=True)['href'])
    return anime_linklist
```

```python
def anime_website_scraper(anime_link):
    import requests
    from bs4 import BeautifulSoup
    anime_web = requests.get(anime_link).text
    soup_ = BeautifulSoup(anime_web,'lxml')
    emt = {}
    emt['Name']= soup_.find('div', itemprop= 'name').strong.text
    emt['Ranked'] = int(soup_.find('span', class_= 'numbers ranked').strong.text[1:])
    emt['Score'] = float(soup_.find('div', class_ = 'fl-l score').div.text)
    emt['Popularity'] = int(soup_.find('span', class_ = 'numbers popularity').strong.text[1:])
    space_it = soup_.find_all('div', class_ = 'spaceit_pad')
    for _ in space_it:
        key = _.text.strip().split(':')[0]
        if key == 'Members':
            break
        if key!= 'Score' and key!= 'Ranked' and key!='Popularity':
            word = _.text.strip().split(':')[1].strip()
            if key == 'Genres' or key == 'Demographic' or key=='Theme' or key=='Themes' or key == 'Genre':
                genre_list = _.find_all('a')
                word = ''
                for count in range(len(genre_list)) :
                    if count==0:
                        word = word + genre_list[count].text
                    else:
                        word = word + ", " + genre_list[count].text
            emt[key] = word.strip().replace('\n','') 
    emt['description'] = soup_.find('p', itemprop = 'description').text.strip().replace('\n','').replace('\r',"")
    return emt
```

```python
%%time
number = int(input("Enter the number of topanime you want to scrape : "))
webpages = get_webpages(number)
empty = []
for webpage_link in webpages:
    print(f"Scraping the {webpage_link}:")
    anime_links = get_anime_links(webpage_link)
    for anime_link in anime_links:
        print(f"Scraping the {anime_link}....")
        empty.append(anime_website_scraper(anime_link))
df = pd.DataFrame(empty).drop(['French','German', 'Spanish'],axis = 1)
df.to_csv(f'top {number} anime.csv')
```
